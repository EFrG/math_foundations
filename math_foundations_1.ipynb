Vectors represent a magnitude and direction from origin.

Norms are functions that quantify vector magnitude (The lenght of a given vector).

#### $L²$ Norm

Of all of the norms, the most important is $L²$ Norm.
\begin{align}
\|\boldsymbol{x}\|_{2} = \sqrt{\sum{i}x_{i}²}
\end{align}
(We sum all of the squared elements of the vector, and then we take the square root out of that).

It measures the simple (Euclidean) distance from origin.

However many dimensions are, vectors in the origin is always just a 0 vector, consiting uniquely of 0s.

$L²$ Norm is the most common norm in machine learning. Instead of $\|\boldsymbol{x}\|_{2}$ it can be noted just as $\|\boldsymbol{x}\|$.

```python
x = np.array([25,2,5])
```

\begin{align}
\|\boldsymbol{x}\|_{2} = \sqrt{\sum{i}x_{i}²}
\end{align}

```python
(25**2 + 2**2 + 5**2)**(1/2)
```

The obtained result, is the $L²$ norm, the euclidean distance, between the vector $x$ and the origin.

```python
np.linalg.norm(x)
```

#### Aditional normalizations:



##### $L¹ Norm$:

It is described as:
\begin{align}
  \|\boldsymbol{x}\|_{1} = \sum_{i} |x_{i}|
\end{align}

Properties:
- Another common norm in ML.
- Varies lineraly at all locations whether near or far from origin.
- Used whenever difference between zero and non-zero is key.

```python
x
```

```python
result = 0
for i in range(len(x)):
  result += np.abs(x[i])
result
```

##### Squared $L²$ norm:

It is expressed as:

\begin{align}
  \|\boldsymbol{x}\|^2_{2} = \sum_{i} x^2 _{i}
\end{align}
Characteristics:

- It is computationally cheaper than $L^2$.
- Squared $L^2$ norm equals simply $\boldsymbol{x}^t \boldsymbol{x}$
- Derivative (used to train many ML algorithms) of element x requires that element alone. hereas $L^2$ norm requires $\boldsymbol{x}$ vector.


- Downside: It grows slowly near origin, so can't be used if distinguishng between zero and near-zero is important.


```python
x
```

```python
(25**2 + 2**2 + 5**2)
```

```python
np.dot(x,x)
```

##### Max Norm (or $L^∞$ norm)
\begin{align}
  \|\boldsymbol{x}\|_{∞} = max_{i}|x_{i}|
\end{align}

- Returns the absolute value of the largest magnitude element.

```python
x
```

```python
np.max(x)
```

##### Generalized $L^p$ Norm:

\begin{align}
  \|\boldsymbol{x}\|_{p} = (\sum_{i}|x_{i}|^p)^\frac{1}{p}
\end{align}

- All normalizations are instances of this generalized $L^p$
- $p\in \mathbb{R}$
- $p \geq 1$

### Basis, Orthogonal, and Orthonormal Vectors.

#### Basis Vectors:

- Can be scaled to represent any vector in a given vector space.

- Typically use unit vectors along axes of vector space.

image.png

#### Orthogonal vectors:

- Two vectors, $\boldsymbol{x}$ , $\boldsymbol{y}$ are orthogonal if:
\begin{align}
  \boldsymbol{x}^T \boldsymbol{y} = 0
\end{align}

- Are at 90º angle to each other, they are orthogonal.

- In an $n$ dimensional space, has, at maximum, $n$ mutually orthogonal vectors.

- **Orthonormal** vectors are orthogonal and all have unit norm. (Like the basis vectors covered previously).
